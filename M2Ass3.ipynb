# Q1

Line plots: Seaborn can be used to create line plots using the lineplot() function. Line plots are useful for visualizing the trend of a continuous variable over time. They can be used to plot time series data or to compare the changes in a variable over different conditions.

Scatter plots: Seaborn provides the scatterplot() function to create scatter plots. Scatter plots are used to visualize the relationship between two continuous variables. They can be used to detect patterns, correlations, and outliers in the data.

Histograms: Seaborn can be used to create histograms using the histplot() function. Histograms are useful for visualizing the distribution of a single continuous variable. They can be used to identify the central tendency, spread, and skewness of the data.

Bar plots: Seaborn provides the barplot() function to create bar plots. Bar plots are used to visualize the relationship between a categorical variable and a continuous variable. They can be used to compare the values of a variable across different categories.

Heatmaps: Seaborn can be used to create heatmaps using the heatmap() function. Heatmaps are useful for visualizing the relationship between two categorical variables or between a categorical variable and a continuous variable. They can be used to identify patterns and correlations in the data.

# Q2

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
sns.scatterplot(x=fmri.timepoint,y=fmri.signal)
```

# Q3

```python
titanic = sns.load_dataset("titanic")
sns.boxplot(x='pclass',y='age',data=titanic)
sns.boxplot(x='pclass',y='fare',data=titanic)
```

# Q4

```python
dia = sns.load_dataset('diamonds')
sns.histplot(data=dia, x='price', hue='cut', multiple='stack')
```

# Q5

```python
far = sns.load_dataset("iris")
sns.pairplot(data=far, hue"species")
```

# Q6

```python
flights = sns.load_dataset('flights')

# Reshape the data to create a pivot table
flights_pivot = flights.pivot('month', 'year', 'passengers')

# Plot a heatmap for the pivot table
sns.heatmap(data=flights_pivot, cmap='Blues', annot=True, fmt='d')
```
