# Q3

```python
import plotly.express as px
tip = px.data.tips()
gir = px.histogram(tip,x="sex",y="total_bill", color="day",pattern_shape="smoker", marginal="box")
gir.update_layout(
    title="Histogram of Total Bill by Gender, Day, and Smoking Status",
    xaxis_title="Gender",
    yaxis_title="Total Bill",
    legend_title="Day",
    font=dict(size=14),
    height=600,
    width=800,
)
gir.show()
```

# Q4

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Load the iris dataset into a Pandas DataFrame
from sklearn.datasets import load_iris
iris_data = load_iris()
iris_df = px.data.iris()

# Create a scatter matrix plot with Plotly Express
fig = px.scatter_matrix(
    iris_df,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species"
)

# Set title for each scatter plot
for i in range(4):
    for j in range(4):
        fig.update_xaxes(title_text=iris_df.columns[i], row=i+1, col=j+1)
        fig.update_yaxes(title_text=iris_df.columns[j], row=i+1, col=j+1)

# Update the figure layout
fig.update_layout(
    height=800,
    width=800,
    title_text="Scatter Matrix Plot of Iris Dataset",
    showlegend=True,
)

# Show the plot
fig.show()
```
